```python
#MA
def MA(r,close):
 
  data_size = len(close)
  # print(data_size)
  TMP = [0 for i in range(data_size)]

  for i in range(r-1,data_size):
    SUM = close[i-(r-1):i].sum()
    TMP[i] = SUM/r
  
  Tmp_dict = {str(r)+'MA':TMP}
  MADF = pd.DataFrame(Tmp_dict)
 
  return MADF
 
# MA(20,Close)
```

```python
#EMA
def EMA(r,close):
  c = 2/(r+1)
  data_size = len(close)    
  TMP = [0 for i in range(data_size)] 
  MADF = MA(r,close)
  mal = MADF[str(r)+'MA'].to_list()
  
  xema = mal[r-1]
  TMP[r] = mal[r-1]
 
  for i in range(r+1,data_size):
    TMP[i] = xema + c*(close[i]-xema)
    xema = TMP[i]
 
  EMAdict = {str(r)+'EMA':TMP}
  EMADF = pd.DataFrame(EMAdict)
  return EMADF
 
# EMA(20,Close)
```

```python
#MACD
def MACD(close):
    data_size = len(close)    
    MACD = []
    MACD_Sig = [0 for i in range(data_size)]
    MACD = pd.DataFrame(MACD)
    MACD['MACD'] = EMA(12,close)['12EMA'] - EMA(26,close)['26EMA']
    MACD_List = MACD['MACD']
    for i in range (34,data_size):
      SUM = MACD_List[i-9:i].sum()
      MACD_Sig[i]= SUM/9
    MACD['MACD_Sig'] = MACD_Sig
    MACD['MACD_OSC'] = MACD['MACD'] - MACD['MACD_Sig']
 
    return MACD
 
# EMA(12,Close)
# MACD(Close)
```

```python
#CCI
 
#판다스 데이터프레임 생략 없이 출력
# pd.set_option('display.max_rows', None)
 
def CCI(close,high,low):
  #CCI_M
  CCI_M = ((close+high+low)/3)
  #print(CCI_M)
 
  #CCI_m
  CCI_m = MA(20,CCI_M)
  CCI_m.rename(columns = {'20MA' : 'CCI_m'}, inplace = True)
  CCI_m = CCI_m.squeeze() 
  # print(CCI_m)
 
  #CCI_dv
  CCI_dv = abs(CCI_M - CCI_m)
  # print(CCI_dv)
 
  #CCI_d
  CCI_d = MA(20,CCI_dv)
  CCI_d.rename(columns = {'20MA' : 'CCI_d'}, inplace = True)
  CCI_d = CCI_d.squeeze() 
  # print(CCI_d)
 
  #CCI
  CCI = (CCI_M - CCI_m)/(CCI_d * 0.015)
  # print(CCI)
  CCI = CCI.to_frame(name = 'CCI')
 
  return CCI
 
 
# CCI(Close,High,Low)
```

```python
#스토캐스틱 12 5 5

def Stochastic(high,low,close):
  #데이터 크기설정
  data_size = len(close)

  #기간 변수 설정
  A = 12
  B = 5
  C = 5

  #변수선언
  Fast_U = [0.0 for i in range(data_size)]
  Fast_L = [0.0 for i in range(data_size)]

  Fast_K = [0.0 for i in range(data_size)]
  Slow_K = [0.0 for i in range(data_size)]
  Slow_D = [0.0 for i in range(data_size)]


  #Fast%K
  for i in range(A,data_size):
    Fast_U[i] = close[i]-min(low[i-A:i])
    Fast_L[i] = max(high[i-A:i])-min(low[i-A:i])
    Fast_K[i] = (close[i]-min(low[i-A:i]))/(max(high[i-A:i])-min(low[i-A:i]))*100
  
  #Fast%D(Slow%K)
  for i in range(A+B,data_size):
    Slow_K[i] = (sum(Fast_U[i-2:i])/3)/(sum(Fast_L[i-2:i])/3)*100
  
  #Slow%D
  for i in range(A+B+C,data_size):
    Slow_D[i] = sum(Slow_K[i-2:i])/3
  
  StoDF = []
  StoDF = pd.DataFrame(StoDF)
  StoDF['Fast%K'] = Fast_K
  StoDF['Slow%K'] = Slow_K
  StoDF['Slow%D'] = Slow_D
  
  return StoDF

Stochastic(High,Low,Close)
#음수가 나오면 안되는데 나옴 나중에 한번 더 검토할 것
#0이 나올리가 없는데 나오는걸 보니 그냥 예시 데이터값에 문제가 있는듯

```

```python
#일목균형표
 
def ILMOK(high,low,close):
 
  data_size = len(close)
 
  Changing = [0.0 for i in range(data_size)]
  Axis = [0.0 for i in range(data_size)]
  Span_A = [0.0 for i in range(data_size)]
  Span_B = [0.0 for i in range(data_size)]
  IM_upper = [0.0 for i in range(data_size)]
  IM_lower = [0.0 for i in range(data_size)]
 
  # #전환선
  # for i in range(10,data_size):
  #   Changing[i] = (max(High[i-9:i])+min(Low[i-9:i]))/2
 
  # #기준선
  # for i in range(27,data_size):
  #   Axis[i] = (max(High[i-26:i])+min(Low[i-26:i]))/2
 
  #선행스팬 A,B
  for a in range(80,data_size):
    i = a-26
    #전환선
    Changing[a] = (max(high[a-9:a])+min(low[a-9:a]))/2
    #기준선
    Axis[a] = (max(high[a-26:a])+min(low[a-26:a]))/2
    #선행스팬 A,B
    Span_A[a] = (((max(high[i-26:i])+min(low[i-26:i]))/2) + ((max(high[i-9:i])+min(low[i-9:i]))/2))/2
    Span_B[a] = ((max(high[i-52:i])+min(low[i-52:i]))/2)
 
  #구름대 윗선 아랫선
  for i in range(80,data_size):
    A = Span_A[i]
    B = Span_B[i]
    if A >= B:
      IM_upper[i] = A
      IM_lower[i] = B
    else:
      IM_upper[i] = B
      IM_lower[i] = A
 
  ilmokDF = []
  ilmokDF = pd.DataFrame(ilmokDF)
  ilmokDF['Changing'] = Changing
  ilmokDF['Axis'] = Axis
  ilmokDF['Span_A'] = Span_A
  ilmokDF['Span_B'] = Span_B
  ilmokDF['IM_upper'] = IM_upper
  ilmokDF['IM_lower'] = IM_lower
 
  return ilmokDF
 
 
ILMOK(High,Low,Close)
```

```python
#RSI
 
#list = [0 for i in range(n)]
#초기 입력 데이터는 close[0] ~ close[data_size-1]의 data_size개 종가임
 
def RSI(close):
 
  N = 14 #권장값
  M = 6  #권장값
  data_size = len(close)
  
  #U, D, AU, AD, RS, RSI, RSI_sig 변수 선언(리스트로)
 
  U = []
  D = []
  AU = []
  AD = []
  RS = []
  RSI = []
  RSI_sig = []
 
  for i in range(data_size):
    U.append(0)
    D.append(0)
    AU.append(0)
    AD.append(0)
    RS.append(0)
    RSI.append(0)
    RSI_sig.append(0)
 
#-------------------   U, D 계산   --------------------
 
  for i in range(data_size-1):
 
    if close[i+1] > close[i]:
      U[i+1] = close[i+1] - close[i]
      D[i+1] = 0
 
    else:
      U[i+1] = 0
      D[i+1] = close[i] - close[i+1]
 
#---------------   AU, AD, RS, RSI 계산   ----------------
 
  for i in range(N-1,data_size):
 
    U_sum = 0
    D_sum = 0
 
    for j in range(0,N):
      U_sum += U[i-j]
      D_sum += D[i-j]
 
    AU[i] = float(U_sum) / N
    AD[i] = float(D_sum) / N
 
    RS[i] = AU[i] / AD[i]
    RSI[i] = RS[i] / (RS[i] + 1) *100
 
# N개 값의 평균이므로 앞의 N-1개 값은 버려짐
 
# 즉, 네 개 변수 모두 앞의 N-1개 값은 0이 들어감
 
# AU[0]부터 AU[N-2]에는 0이
# AU[N-1]부터는 이전 N개의 평균값이 들어감
 
#------------------   RSI_sig 계산   -------------------
 
  for i in range(N+M-2,data_size):
    
    RSI_sum = 0.0
    for j in range(0,M):
      RSI_sum += RSI[i-j]
 
    RSI_sig[i] = RSI_sum / M
    
# N개 값의 평균의 M개 값의 평균이므로 앞의 N+M-2개 값은 버려짐
 
#-------------------------------------------------------
#RSI 시그널과 rsi df형태로 리턴
 
  Idict = {'RSI':RSI,'RSI_sig':RSI_sig}
  LDF = pd.DataFrame(Idict)
  return LDF
 
# RSI(Close)
```

```python
#투자 심리도
def Inv_desire(close):
  data_size = len(close)
  UDDF = [0.0 for i in range(data_size)]
  UDDFP = [0.0 for i in range(data_size)]  
  for i in range(1,data_size):
  
    A = close[i-1]
    B = close[i]
    if A>=B:
      UDDF[i] = 0
    else:
      UDDF[i] = 1
  
  for i in range(10,data_size):
    UDDFP[i] = sum(UDDF[i-9:i])*10
  UDDFP = pd.DataFrame(UDDFP,columns = ['Inv_D'])
  #UDDFP.rename(columns = ,inplace = True)
  return UDDFP
 
#Inv_desire(Close)
```
